```python
# find_elements 取值

driver = webdriver.Chrome(service=Service())
driver.get("https://goodinfo.tw/tw/ShowK_Chart.asp?STOCK_ID=%E5%8A%A0%E6%AC%8A%E6%8C%87%E6%95%B8&CHT_CAT=DATE32") # 瀏覽股票網站

time.sleep(5) # 等待10秒 讓頁面有足夠時間，讓元素或資料載入

#get_attribute() 抓屬性

trs = driver.find_element(By.CSS_SELECTOR , '.row_bg_2n.row_mouse_over > tbody').find_elements(By.TAG_NAME , 'tr') 

# print(trs)

for tr in trs:
    td = tr.find_element(By.CSS_SELECTOR ,'td:nth-child(2)')
    print(td.get_attribute("outerHTML"))

    a = td.find_element(By.TAG_NAME,'a')

    print(a.text)
    print(a.get_attribute("href"))
    print(a.get_attribute("class"))
    print(a.get_attribute("tag_name"))
    print(a.is_displayed()) # 是否顯示在畫面
    print(a.location) # 畫面的位置
    print(a.is_enabled()) # 是否可被互動 // 是否可被點 ? 
    print(a.is_selected()) # 是否被選取 ? checkbox 


          
    print('--------------------------------------------------------------------------------')

# 停止程式
driver.close()
driver.quit()
```

```python
# 等待時間設定 1 

driver = webdriver.Chrome(service=Service())
driver.get("https://goodinfo.tw/tw/ShowK_Chart.asp?STOCK_ID=%E5%8A%A0%E6%AC%8A%E6%8C%87%E6%95%B8&CHT_CAT=DATE32") # 瀏覽股票網站

# time.sleep(5) # 不推薦


# 全域設定等待時間 implicitly_wait 

# 如果等了幾秒 還是沒出現就報錯

# 優點 : 設定簡單 ，  設定一次即可使用 !

# 缺點 : 因為統一時間，無法客製化


driver.implicitly_wait(10) # 每個元素最多等待 10 秒


driver.find_element(By.ID , 'ats-interstitial-button').click() # 去除廣告

html_input = driver.find_element(By.ID , 'txtStockCode')

html_input.send_keys('2330') # 輸入值

driver.find_element(By.ID , 'btnStockSearch').click() # 點擊按鈕

#notExistTag = driver.find_element(By.ID , 'ABC123')

time.sleep(10) # 觀看用

# 停止程式
driver.close()
driver.quit()
```

```python
driver = webdriver.Chrome(service=Service())

driver.get("https://www.google.com/")

time.sleep(1)

driver.get("https://khh.travel/zh-tw/attractions/list/") 

time.sleep(2) # 等兩秒後開始

driver.implicitly_wait(10) # 每個元素最多等待 10 秒


data = []
exist = []

while len(data) < 100 :

    
    driver.execute_script(f"window.scrollBy(0, 300);")

    time.sleep(0.1)
    
    # 檢查是否已到底部：可視高度 + 當前滾動位置 是否大於或等於頁面總高度
    at_bottom = driver.execute_script("return window.innerHeight + window.scrollY >= document.documentElement.scrollHeight;")
    if at_bottom:
        print("已到達頁面底部，停止滾動")
        break

   
    ul = driver.find_element(By.CSS_SELECTOR , '.row.g-2.g-md-3.g-xl-4')


    ul = BeautifulSoup(ul.get_attribute('outerHTML'),'html.parser') 

    lis = ul.select("li")

    for li in lis :
        # print(li)
        name = li.find(class_ = 'mb-1 mb-md-2 fz-18px fz-xl-20px font-weight-bold lh-initial line-clamp-2').text
        
        OpeningHours = li.find(class_ = 'fz-14px fz-xl-16px text-success text-truncate')
        if OpeningHours != None :
            OpeningHours = OpeningHours.text
        else:
            OpeningHours = ''
            
        popularity = li.find(class_ = 'd-flex align-items-center ml-auto text-info fz-13px fz-xl-15px')
        if popularity != None :
            popularity = int(popularity.text[2:])
        else:
            popularity = 0
            
      
        if name not in exist:
            # print(name)
            # print(OpeningHours)
            # print(popularity )
            # print('-----')
            exist.append(name)
            data.append({
                "景點名稱" : name ,     
                "開放時間" : OpeningHours ,                
                "受歡迎程度" : popularity ,                
            })
            

import pandas as pd
df = pd.DataFrame(data)
# df.to_csv('viewpoint2.csv' , encoding='utf-8' )
df.head(100)   
```
