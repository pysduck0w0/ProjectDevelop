```python
#UBIKE 網站
# https://www.youbike.com.tw/region/main/stations/list/

import requests
import pandas as pd

url = "https://apis.youbike.com.tw/json/area-all.json"

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
data[0]
```

```python
df = pd.DataFrame(data)

df.head(100)
```

```python
city_code = df[["area_code","area_code_2","area_english","area_name_tw","is_open","register_card","station_start","station_end"]]
city_code.head(100)
```

```python
print(city_code["area_name_tw"].is_unique)
print(city_code["area_code"].is_unique)
print(city_code["area_code_2"].is_unique)
```

```python
city_code = city_code.sort_values(by = ["area_code_2"])

city_code
```

```python
register_card_df = df[["area_code","area_code_2","area_name_tw","register_card"]]
register_card_df.head(100)
```

```python
taipei_card = register_card_df[register_card_df["area_name_tw"] == "台北市"]
taipei_card =  taipei_card.iloc[0]
taipei_card = taipei_card["register_card"]
taipei_card 
```

```python
taipei_card  = pd.DataFrame(taipei_card ) 
taipei_card.head(10)
```

```python
area_name_tws = city_code["area_name_tw"].unique() 
print(area_name_tws)
```

```python
card_table = None

for area_name_tw in area_name_tws:
    city_register_card = register_card_df[register_card_df["area_name_tw"]==area_name_tw].iloc[0]["register_card"]
    city_register_card = pd.DataFrame(city_register_card)

    print(area_name_tw,city_register_card["name_tw"].unique())

    if card_table is None :
        card_table = city_register_card
    else:
        card_table.merge(city_register_card , on = "card_type",how = "left")

card_table.head(100)   
```

```python
#查詢各縣市 ubike 站點數量

headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36",
    "Referer": "https://www.youbike.com.tw/",
}

url = 'https://apis.youbike.com.tw/api/front/station/count'

response = requests.get(url , headers = headers)
#response = requests.get(url)

data = []

if response.status_code == 200:
    data = response.json()
    print(data)
    if data["retMsg"] == "查詢各縣市場站數量成功" :
        data = data["retVal"]
else:
    print(f"HTTP {response.status_code}:{response}")
data = pd.DataFrame(data)
data.head()
```

```python
city_code = city_code.merge(data , on = "area_code",how = "left")

city_code.head(20)
```

```python
# 查詢全台所有 ubile 站點

headers = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36",
    "Referer": "https://www.youbike.com.tw/",
}

url = 'https://apis.youbike.com.tw/json/station-yb2.json'

response = requests.get(url , headers=headers)

data = []

if response.status_code == 200:
    data = response.json()
   
else:
    print(f"HTTP {response.status_code}: {response}")

df = pd.DataFrame(data)
df.head()
```
