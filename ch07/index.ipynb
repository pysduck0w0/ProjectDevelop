```python
#UBIKE 網站
# https://www.youbike.com.tw/region/main/stations/list/

import requests
import pandas as pd

url = "https://apis.youbike.com.tw/json/area-all.json"

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
data[0]
```

```python
df = pd.DataFrame(data)

df.head(100)
```

```python
city_code = df[["area_code","area_code_2","area_english","area_name_tw","is_open","register_card","station_start","station_end"]]
city_code.head(100)
```

```python
print(city_code["area_name_tw"].is_unique)
print(city_code["area_code"].is_unique)
print(city_code["area_code_2"].is_unique)
```

```python
city_code = city_code.sort_values(by = ["area_code_2"])

city_code
```

```python
register_card_df = df[["area_code","area_code_2","area_name_tw","register_card"]]
register_card_df.head(100)
```

```python
taipei_card = register_card_df[register_card_df["area_name_tw"] == "台北市"]
taipei_card =  taipei_card.iloc[0]
taipei_card = taipei_card["register_card"]
taipei_card 
```
