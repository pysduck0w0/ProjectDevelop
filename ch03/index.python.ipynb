```python
# =================== 小測驗 ===============================


import pandas as pd

data = {
    '姓名': ['小明', '小華', '小美', '小強', '小美芳', '小陳', '小李', '小趙', '小王', '小張'],
    '性別': ['男', '男', '女', '男', '女', '男', '男', '女', '男', '女'],
    '年齡': [15, 16, 15, 17, 16, 15, 17, 16, 15, 17],
    '身高': [170, 165, 160, 175, 158, 168, 180, 162, 172, 169],  # 單位：公分
    '體重': [60, 58, 50, 75, 52, 65, 75, 81, 65, 65]           # 單位：公斤
}

df = pd.DataFrame(data)
df.head()


# 題目 1：計算 BMI 並新增 BMI 欄位

# 公式 : BMI = 體重 / ( 身高 / 100 ) ** 2

# 題目 2：塞選男性過重學生 BMI >= 24

# 題目 3：依照性別與身高做排序 升序

# 題目 4：請統計各年齡數量

# 題目 5：請統計最高最低身高與平均身高 
```

```python
#1
df["BMI"] = df.apply(lambda row : row["體重"] / (row["身高"] / 100)**2 ,axis = 1)

df.head()
```

```python
#2
results = df[(df["BMI"]>= 24) & (df["性別"] =="男")]
results.head()
```

```python
#3 ascending= False 遞減
sort_results = df.sort_values(by = ["性別","身高"])
sort_results.head(10)
```
