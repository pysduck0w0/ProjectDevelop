```python
import pandas as pd
```

```python
sr = pd.Series([60,81,100,73,50,66])
print(sr)
```

```python
indexs = sr.index
print(indexs)

values = sr.values
print(values)
```

```python
sr = pd.Series([60,90,55,70,30,None],index =["Sofia","Jack","Peter","Dan","Rita","Steven"])
print(sr)
```

```python
index = sr.index
print("抓取名子",index)
```

```python
result = sr[sr>=60]
print("及格名單")
print(result)
```

```python
results = sr + 10
print("加分結果")
print(results)

results = sr - 10
print("扣分結果")
print(results)

results = sr * 1.1
print("乘10%結果")
print(results)

results = sr / 2
print("除2結果")
print(results)

```

```python
if "Sofia"  in sr:
    print("Sofia存在")
if "Sofia" not in sr:
    print("Sofia不存在")
```

```python
print("判斷誰有空值")
isNull = pd.isnull(sr)
print(isNull)

print("判斷誰沒有空值")
isNotNull = pd.notnull(sr)
print(isNotNull)

isNull = sr.isnull()
print(isNull)
```

```python
sr2 = pd.Series([49,80],index = ["Grace","Kity"])
sr3 = pd.Series([30,95],index = ["Grace","Kity"])

sr4 = sr2 + sr3
print(sr4)

sr4 = sr2 - sr3
print(sr4)

sr4 = sr2 * sr3
print(sr4)

sr4 = sr2 / sr3
print(sr4)

```

```python
sr.name = " 分數"
sr.index.name ="姓名"
print("設定名稱後的sr:")
print(sr)
```

```python
print("Rita與Steven的分數:")
print(sr.loc[["Rita","Steven"]])
print()
```

```python
print("第一筆和第三筆的資料:")
print(sr.iloc[[0,2]])
```

```python
print("第一筆到第四筆:")
print(sr.iloc[1:5])#range原理
```

```python
def grade(score):

    if pd.isnull(score):
        return "缺考"
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    elif score < 60:
        return "F"
results = sr.map(grade)
print(results)

results = sr.map(lambda score : "及格" if score >= 60    else "不及格")
print(results)
    
```

```python
results = sr.apply(lambda x:100 if (x*1.1+5)>100 else x*1.3+5)
print(results)
```

```python
#統計方式
print("描述性統計")
print(sr.describe())
```

```python
print("依照分數排序(遞增)")
print(sr.sort_values())
print()

print("依照人名排序(遞增)")
print(sr.sort_index())
print()

print("重新設定索引")
print(sr.reset_index())
print()

```

```python
#去除空值
results = sr.dropna()
print(results)
```

```python
#轉型態

results = sr.dropna().astype(int)
print(results)

results = sr.dropna().astype(str)
print(results)

results = sr.dropna().astype(int).astype(str)
print(results)

```
