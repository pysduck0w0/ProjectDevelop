```python
sr = pd.Series([60,90,55,70,30,None],index =["Sofia","Jack","Peter","Dan","Rita","Steven"])
print(sr)
```

```python
def grade(score):

    if pd.isnull(score):
        return "缺考"
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    elif score < 60:
        return "F"
results = sr.map(grade)
print(results)

results = sr.map(lambda score : "及格" if score >= 60    else "不及格")
print(results)
    
```

```python
results = sr.apply(lambda x:100 if (x*1.1+5)>100 else x*1.3+5)
print(results)
```

```python
#統計方式
print("描述性統計")
print(sr.describe())
```

```python
print("依照分數排序(遞增)")
print(sr.sort_values())
print()

print("依照人名排序(遞增)")
print(sr.sort_index())
print()

print("重新設定索引")
print(sr.reset_index())
print()

```

```python
#去除空值
results = sr.dropna()
print(results)
```

```python
#轉型態

results = sr.dropna().astype(int)
print(results)

results = sr.dropna().astype(str)
print(results)

results = sr.dropna().astype(int).astype(str)
print(results)

```

```python
#新增與修改

print(sr)

sr['Sofia'] = 90
sr['Peter'] = 73

print(sr)

sr2 = pd.Series([8,58,73],index = ["Sofia","Jack","Steven"])#建立新sr2
print(sr2)

results = pd.concat([sr,sr2])

print(results)
```

```python
#groupby 

sr2 =  pd.Series([8,58,73],index = ["Sofia","Jack","Steven"])
sr3 = pd.concat([sr,sr2])
sr3 = sr3.dropna()

print(sr3.sort_values())
print(sr3.index)
results = sr3.groupby(sr3.index).mean()
print("重複索引資料以平均聚合")
print(results)

results = sr3.groupby(sr3.index).sum().astype(int)
print("重複索引資料以加總聚合")
print(results)
```

```python
score_data = pd.Series([60,81,100,73,50,None],index = ["小明","小花","麗麗","博明","詩涵","俊雅"])
print(score_data)

#1
print(1)
print(score_data.iloc[1:4])

#2
print(2)
print(score_data.map(lambda score :  "及格" if score >=60 else "不及格"  ))

#3
print(3)
print(score_data.apply(lambda x : 100 if x*1.1 +5 >=100 else x*1.1 +5))

#4

score_data2 = pd.Series([80],index = ["小林"])
print(4)
print(pd.concat([score_data,score_data2]))

#5
print(5)
score_data_1 = pd.Series([60, 81, 100, 73, 50, None], index=['小明', '小花', '麗麗', '博明', '師函', '俊雅'])
score_data_2 = pd.Series([88, 80, 89, 68], index=['小明', '師函', '博明', '麗麗']) 

results = pd.concat([score_data_1,score_data_2])
results = results.dropna() 

print(results)

print(results.groupby(results.index).sum().astype(int))
```
