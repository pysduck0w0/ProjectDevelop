```python
def get_page(url):
        
    html_code = None 
    
    response = requests.get(url)
    
    if response.status_code == 200 :
        
        html_code = BeautifulSoup(response.text,'html.parser') 
        if html_code.find("div" , class_='mb-1 fz-38px fz-md-50px font-weight-bold text-dark mb-3 text-center text-xl-left' )  !=  None:
            html_code = None
        
    return html_code


 
def get_title (html_code):

    h2 = html_code.find("h2", class_ = 'fz-30px fz-mx-32px font-weight-bold')

    if h2 != None:
        return h2.text

    h2 = html_code.find("h2", class_ = 'w-100 max-w-[1200px] mx-auto font-weight-bold fz-40px fz-md-48px fz-xl-64px text-white')

    if h2 != None:
        return h2.text
        
    return None;

def get_description(html_code):

    description = html_code.find( class_ = 'max-w-[800px] mt-2 fz-18px fz-md-20px fz-xl-24px lh-initial line-clamp-2')

    if description != None:
        return description.text
         
    return ''


def get_description_img(html_code):
    
    div_tag = html_code.find( class_ ="manual-script-blk mt-2 mt-md-3 fz-16px fz-xl-18px letter-spacing-1px")

    imgs = div_tag.find_all('img')

    data = []
    
    for img in imgs:

        data.append({
            'img_alt' :  img.get('alt'),
            'img_src' :  img.get('src')
        })
    

    return data

def get_information(html_code):
    
    div_tag = html_code.select( ".py-5.border-bottom > .pt-3.fz-16px.fz-xl-18px > div")

    data = {
        'phone' : None ,
        'address' : None ,
        'googleMap' : None ,
        'openingHours' : None ,
        'ticketInformation': None ,
    }

    
    for i in range(0,len(div_tag)):
        d_flex = div_tag[i]   
        title = d_flex.find('div',class_='font-weight-bold')
        if title == None :
            continue
        else:
            title.get_text()

        if '電話' in title :
            data['phone'] = d_flex.find_all('div')[-1].text
            
        elif'地址' in title :
            a_tag =  d_flex.find('a') 
            if a_tag != None :
                data['address'] = a_tag.find('span').text
                data['googleMap'] = a_tag.get('href')
            
        elif '開放時間' in title :            
            data['openingHours'] = d_flex.find_all('div')[1].get_text()

        elif '門票資訊' in title:
            data['ticketInformation'] = d_flex.find('div' ,class_='manual-script-blk').get_text("\n", strip=True)

    return data 
   

import time 

def main():
        
    json_data = []
    for i in range(1,357):
        # time.sleep(0.5)
        url = f'https://khh.travel/zh-tw/attractions/detail/{i}'
        html_code = get_page(url)
        
        if  html_code == None :
            continue

        print('爬蟲 url' , url)
        results = {
        
            'title' : None ,
            'description' : None ,
            'description_img' : None,
            'information' : None ,
        }

        results['title'] = get_title(html_code)    
        results['description'] = get_description(html_code)
        results['description_img'] = get_description_img(html_code)
        results['information'] = get_information(html_code)
        
        json_data.append(results)
        # print(results)
            
    return json_data
    
    

        
results = main()


results    
    
```
