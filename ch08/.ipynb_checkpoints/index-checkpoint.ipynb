```python
import pandas
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://khh.travel/zh-tw/attractions/detail/2/'


response = requests.get(url)

html_text = None

if response.status_code == 200 :
    html_text = response.text

print(html_text)
```

```python
soup = BeautifulSoup(html_text,"html.parser")

print(soup.prettify)
```

```python
#拿取 Tag
print(soup.title)
#拿取 Tag
print(soup.a)
#拿取標籤文字
print(soup.title.text)
```

```python
#find 尋找第一格符合的標籤
print(soup.find("a"))
#find 用 class 尋找
print(soup.find(class_='fz-30px fz-mx-32px font-weight-bold'))
#find 用 id 尋找
print(soup.find(id='img-18322-1591088705541'))
```

```python
img_tag = soup.find(id='img-18322-1591088705541')

print(img_tag)

print(img_tag.get("alt"))

print(img_tag.get("stytle"))

print(img_tag.get("src"))
```

```python
#剝洋蔥式抓取法

div_tag = soup.find(class_="d-flex justify-content-between justify-content-md-start align-items-center py-2 py-xl-3 border-bottom")
print(div_tag)
print("--------------------------------------------------------------------------------------------------------------------------")

d_flex = div_tag.find(class_="d-flex")

print(d_flex)

release_time = d_flex.find(class_ ="d-none d-md-block text-info fz-13px fz-xl-16px mr-1 mr-md-2").text
update_time = d_flex.find(class_ ="text-info fz-13px fz-xl-16px mr-1 mr-md-2").text
print("--------------------------------------------------------------------------------------------------------------------------")
print(release_time)
print(update_time)
```

```python
#使用 split 切割

results_list = release_time.split(":")

print(results_list)
print(results_list[1])
```

```python
#find_all 推薦用法
list = soup.find_all(class_ ="d-flex align-items-start mb-2 mb-0-last")
print(list)

print(list[0])
```

```python
#使用 for 迴圈 一一撈取 玩樂攻略

for li in list:
    print(li.text)
```

```python
#拿去設施
list = soup.find_all(class_ ="mr-2 mr-0-last d-flex mb-1 align-items-center")
for li in list:
    print(li.text)
```

```python
#搭配 function 過濾

imgs = soup.find_all("img", alt = lambda alt: "崗山之眼" in alt)
print(imgs)
for img in imgs:
    print(img)
    print(img.get("src"))
```
